```python
import zipfile
import json
import re
from pymongo import MongoClient
from bson.objectid import ObjectId

# data downloaded from https://www.kaggle.com/datasets/thechosanone/donald-trumps-tweets
zip_file_path = 'donald_data.json.zip'
json_file_name = 'donald_data.json'
```

## import

```python
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    with zip_ref.open(json_file_name, 'r') as json_file:
        lines = json_file.readlines()

# –û–±—Ä–æ–±–∏—Ç–∏ –∫–æ–∂–µ–Ω —Ä—è–¥–æ–∫ —è–∫ –æ–∫—Ä–µ–º–∏–π JSON –æ–±'—î–∫—Ç
data = []
for line in lines:
    try:
        obj = json.loads(line.strip())
        data.append(obj)
    except json.JSONDecodeError:
        # –ü—Ä–æ–ø—É—Å—Ç–∏—Ç–∏ —Ä—è–¥–∫–∏, —è–∫—ñ –Ω–µ –º–æ–∂–Ω–∞ —Ä–æ–∑–ø–∞—Ä—Å–∏—Ç–∏ —è–∫ JSON
        pass

print("len =",len(data))
data[0]
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['tweets_data']
collection = db['donald`s_tweets']
```

```python
collection.insert_many(data)
```

```python
count = collection.count_documents({})
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç—ñ–≤ —É –∫–æ–ª–µ–∫—Ü—ñ—ó: {count}")
```

## Analyse

```python
field_name = 'place'
unique_values = collection.distinct(field_name)
print(f"–£–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è –ø–æ–ª—è '{field_name}': {len(unique_values)}")
unique_values
```

```python
filter = {
    'place.country': 'United States'
}
cursor = collection.find(filter)
results = []
for document in cursor:
    results.append(document)
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å: ",len(results))
results
```

```python
regex = re.compile("^U",re.IGNORECASE)  
filter = {
    'place.country': 'regex'
}
cursor = collection.find(filter)
results = []
for document in cursor:
    results.append(document)
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å: ",len(results))
results
```
